```python
data = pd.read_excel ('data_cleaning.xlsx', sheet_name='Usage',header = None)
data.head(10)
```

```python
df=data.copy()
df
```

```python
df[0] = df[0].astype(str).str.replace('_', ' ').str.strip()
df
```

```python
Time= []

for i in df[0]:
    time = re.findall("^\d+\ *[A-Z]+", i)
    Time.append(time)
```

```python
Day= []

for i in df[0]:
    day = re.findall("[A-Z]+[a-z]+\s", i)
    Day.append(day)
```

```python
Date=[]
for i in df[0]:
    date = re.findall("\d+[a-z]*\-[A-Z]+[a-z]{2}\-[0-9]{4}", i)
    Date.append(date)
```

```python
Kwh=[]
for i in df[0]:
    kwh = re.findall("[0-9]\.\d+", i)
    Kwh.append(kwh)
```

```python
new_df = zip(Time, Day, Date, Kwh)
column_names = ('time','day', 'date', 'kwh')
df_electric = pd.DataFrame(new_df, columns=column_names)
```

```python
df_electric
```

```python
time = df_electric['time'].apply(pd.Series, index = ['time'])
day = df_electric['day'].apply(pd.Series)
date =  df_electric['date'].apply(pd.Series,  index = ['date'])
kwh =  df_electric['kwh'].apply(pd.Series,  index = ['kwh'])
```

```python
data = pd.concat([time, day, date, kwh], axis=1)
data
```

```python
data['time'] = data['time'].astype(str).str.replace(' ', '')
data
```

```python
data[['Time', 'AM/PM']] = data.time.str.split(r"[A-Z]{2}",expand=True) 
data
```

```python
data[['Time_bosh','AM/PM']] = data.time.str.split(r"\d+",expand=True) 
data
```

```python
data['Time'] = data['Time'].apply(lambda x: x.zfill(2))
data
```

```python
data[['Day','Month', 'Year']] = data.date.str.split("-",expand=True) 
data
```

```python
data['Day'] = data['Day'].astype(str).str.replace(r'[a-z]', '')
data
```

```python
data['Day'] = data['Day'].apply(lambda x: x.zfill(2))
data
```

```python
data['Full_date']=data['Day'].astype(str)+'-'+data['Month']+'-'+data['Year']
data
```

```python
data['Full_date'] = pd.to_datetime(data['Full_date'])
data
```

```python
data['Week_day'] = data['Full_date'].dt.day_name()
data
```

```python
data.drop(columns =['date','time', 'Time_bosh'], inplace = True) 
data
```

```python
data.drop(columns=[0],inplace = True) 
data
```

```python
data = data[['Full_date','Time', 'AM/PM', 'Week_day','Day', 'Month', 'Year', 'kwh']]
data
```

```python
data.columns=data.columns.str.lower()
data
```

```python
data.sort_values(by = ['full_date', 'am/pm', 'time'], inplace = True)
```

```python
data
```

```python
data.to_csv('Usage_kwh.csv', index=False)
```
